```python
from qopt import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# rabi driving in lab frame

n_time_steps = 10
time_step = .1
time_steps = time_step * np.ones(n_time_steps)
time_acc = time_step * np.arange(n_time_steps)
omega_0 = 2 * np.pi
n_opt_pars = 2
n_ctrl_amps = 1

def amp_func(opt_pars):
    ctrl_amps = np.empty((opt_pars.shape[0], n_ctrl_amps))
    ctrl_amps[:, 0] = opt_pars[:, 0] * np.sin(omega_0 * time_acc + opt_pars[:, 1])
    return ctrl_amps

def deriv_amp_func(opt_pars):
    ctrl_amps_deriv = np.empty((opt_pars.shape[0], n_opt_pars, n_ctrl_amps))
    ctrl_amps_deriv[:, 0, 0] = np.sin(omega_0 * time_acc + opt_pars[:, 1])
    ctrl_amps_deriv[:, 1, 0] = opt_pars[:, 0] * np.cos(omega_0 * time_acc + opt_pars[:, 1])
    return ctrl_amps_deriv

amplitude_function = CustomAmpFunc(
    value_function=amp_func,
    derivative_function=deriv_amp_func
)
```

```python
example_drive = np.ones((10, 2))
example_drive[:, 1] = 0
plt.plot(amplitude_function(example_drive))
plt.title('Driving Amplitude')
plt.xlabel('Time (a.u.)')
plt.ylabel('Control Amplitude')
plt.savefig(r'H:\SimulationData\GroupSeminar\amp_func.png')
```

### Transfer Function

```python
np.random.seed(0)

transfer_func = ExponentialTF(
    awg_rise_time=.02,
    oversampling=10,
    num_ctrls=1
)
transfer_func.set_times(time_step * np.ones((10, 1)))
transfer_func.plot_pulse(np.random.rand(10, 1)-.5)
```

### Solver Algorithm

```python
solver_lab_frame = SchroedingerSolver(
    h_ctrl=[.5 * DenseOperator.pauli_x()],
    h_drift=[omega_0 * .5 * DenseOperator.pauli_z()],
    tau=time_steps,
    transfer_function=transfer_func,
    amplitude_function=amplitude_function
)
```

```python
solver_rotating_frame = SchroedingerSolver(
    h_ctrl=[.5 * DenseOperator.pauli_x(), .5 * DenseOperator.pauli_x()],
    h_drift=[0 * DenseOperator.pauli_z()],
    tau=time_steps,
    transfer_function=transfer_func,
    amplitude_function=amplitude_function
)
```

```python
n_time_steps = 200
time_step = .05
time_steps = time_step * np.ones(n_time_steps)

ntg_quasi_static = NTGQuasiStatic(
    n_samples_per_trace=n_time_steps,
    standard_deviation=[.3],
    n_traces=100,
    always_redraw_samples=False,
    sampling_mode='uncorrelated_deterministic'
)

qs_noise_solver = SchroedingerSMonteCarlo(
    h_ctrl=[.5 * DenseOperator.pauli_x(),
            .5 * DenseOperator.pauli_y()],
    h_drift=[0 * DenseOperator.pauli_z()],
    h_noise=[.5 * DenseOperator.pauli_x()],
    tau=time_steps,
    noise_trace_generator=ntg_quasi_static,
    processes=5
)

fidelities = np.empty(1001, dtype=np.complex128)

ctrl_amps = 5 * np.ones((n_time_steps, 2))
ctrl_amps[:, 1 ] = 0
qs_noise_solver.set_optimization_parameters(ctrl_amps)
up = DenseOperator(np.asarray([[1], [0]]))
for propagators_per_trace in qs_noise_solver.forward_propagators_noise:
    for i, propagator in enumerate(propagators_per_trace):
        fidelities[i] = state_fidelity(up.dag(), propagator * up)

# less pythonic way to make it more readable
```

```python
propagated_states = [[propagator * up for propagator in prop_trace]
                     for prop_trace in qs_noise_solver.forward_propagators_noise]
fidelities = np.asarray([[state_fidelity(up.dag(), prop_state) for prop_state in state_trace]
              for state_trace in propagated_states])
```

```python
probabilities = np.asarray(
    [[np.real(prop_state[0, 0]) for prop_state in state_trace]
     for state_trace in propagated_states]
)
```

# Optimization

```python
n_time_steps = 20
time_step = .1
time_steps = time_step * np.ones(n_time_steps)

ntg_quasi_static = NTGQuasiStatic(
    n_samples_per_trace=n_time_steps,
    standard_deviation=[.2],
    n_traces=6,
    always_redraw_samples=False,
    sampling_mode='uncorrelated_deterministic'
)

qs_noise_solver = SchroedingerSMonteCarlo(
    h_ctrl=[.5 * DenseOperator.pauli_x(),
            .5 * DenseOperator.pauli_y()],
    h_drift=[0 * DenseOperator.pauli_z()],
    h_noise=[.5 * DenseOperator.pauli_z()],
    tau=time_steps,
    noise_trace_generator=ntg_quasi_static
)

x_pi_half = (.5 * DenseOperator.pauli_x()).exp(1j * .5 * np.pi)
entanglement_infid = OperationInfidelity(
    solver=qs_noise_solver,
    target=x_pi_half
)
qs_noise_infid = OperationNoiseInfidelity(
    solver=qs_noise_solver,
    target=x_pi_half,
    neglect_systematic_errors=True
)

simulator = Simulator(
    solvers=[qs_noise_solver, ],
    cost_fktns=[entanglement_infid, qs_noise_infid]
)

bounds = [-2 * np.pi * np.ones((n_time_steps * 2)),
          2 * np.pi * np.ones((n_time_steps * 2))]
optimizer = LeastSquaresOptimizer(
    system_simulator=simulator,
    bounds=bounds,
    cost_fktn_weights=[1, 5e2]
)

np.random.seed(0)
initial_pulse = np.random.rand(n_time_steps, 2)
result = optimizer.run_optimization(initial_pulse)
data = DataContainer()
data.append_optim_result(result)
analyser = Analyser(data)
analyser.plot_costs()
```
